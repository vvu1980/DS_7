```python
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

                    
```

```python
#рохраховує мінімальне значення
def min_numbers(list_with_random_numbers):
    i = 0
    while i <= len(list_with_random_numbers)-1:
        if list_with_random_numbers[i] < list_with_random_numbers[i-1]:
            min_numbers = list_with_random_numbers[i]
        i+=1
    return min_numbers
print(min_numbers)
```

```python
# медіану послідовності
def median_num(a):
    a.sort()
    if len(a) % 2 != 0:
        index = int(((len(a)-1)/2)+1)
        return a[index]
    else:
        index = int(((len(a)-1)/2))
        return (a[index]+a[index+1])/2

    
```
