**Створіть нову колонку "Actual price", яка утворюється із "Purchase Price" та націнки за принципом:**
- якщо "Purchase Price" > 50, націнка 20%
- якщо "Purchase Price" > 100, націнка 10%
- в інших випадках націнка 30%
   

Составляем функцию с условием наценки для Actual price

```python
def f(row):
    if row['Purchase Price'] > 100:
        val = row['Purchase Price'] * 1.2
    elif 50 < row['Purchase Price'] <= 100:
        val = row['Purchase Price'] * 1.1
    else:
        val = row['Purchase Price'] * 1.3
    return val
```

Создаем новую колонку Actual price, применяя функцию к df

```python
df['Actual price'] = df.apply(f, axis=1)
```

Дополняем колонкой с коеффициентом наценки

```python
df['Up_price'] = df['Actual price']/df['Purchase Price']
```

Выводим новые колонки

```python
df.loc[:, ['Purchase Price', 'Actual price', 'Up_price']]
```
