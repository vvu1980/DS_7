**Формируем цикл для перебора ссылок по дате. Сохраняем данные в списко cur_db**

```python
import requests
from datetime import datetime, timedelta
d = datetime.today()

cur_db = []

for i in range(365):
    site = f'https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?valcode=USD&date={d.strftime("%Y%m%d")}&json'
    r = requests.get(site)
    if r.status_code == 200:
        data = r.json()
        cur_db.append(data)
    d -= timedelta(days=1)

cur_db
```

**Поскольку данные сложились в словари внутри списка, который внутри большого списка, упрощаем задачу и формируем временный списко temp_l для облегчения обработки скаченных данных. На выходе получаем списко словарей по дате**

**Убираем лишние столбцы, а столбец exchangedate меняем тип данных на Дату**

```python
temp_l = []
for el in cur_db:
    for dic in el:
        print(dic)
        temp_l.append(dic)
for row in temp_l:
    del row['r030']
    del row['txt']
    del row['cc']
    d = datetime.strptime(row['exchangedate'],"%d.%m.%Y")
    row['exchangedate'] = d
temp_l
```

**Делаем запрос к сайту**

```python
import requests
from bs4 import BeautifulSoup

HEADERS = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:49.0) Gecko/20100101 Firefox/49.0',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
            'Accept-Language': 'en-US,en;q=0.5',
            'Accept-Encoding': 'gzip, deflate',
            'DNT': '1',
            'Connection': 'keep-alive',
            'Upgrade-Insecure-Requests': '1'
        }

SITE = 'https://smallpacking.agrosem.ua/shop/'
r = requests.get(SITE, HEADERS)
r.status_code
```

**Подбираем объекты поиска**

```python
#Название продукта

soup = BeautifulSoup(r.text, 'lxml')
product = soup.find('div', class_="product-from-category-container")
product.span.text
```

**Переходим внутрь товара для поиска специфических элементов**

```python
link = product.a.get('href')
link
```

```python
#Цена

def get_price(link):
    ''' Переходим на страницу конкретного товара и ищем ЦЕНУ товара'''
    
    r = requests.get(link, HEADERS)
    soup = BeautifulSoup(r.text, 'lxml')
    p = soup.find('span', class_="regular-price").text
    q = len(p)
    pr = int(''.join(p[0:len(p)-4:1].split()[0:2]))
    
    return pr
```

```python
#Код

def get_i_kod(link):
    ''' Переходим на страницу конкретного товара и ищем КОД товара'''
    
    r = requests.get(link, HEADERS)
    soup = BeautifulSoup(r.text, 'lxml')
    i = soup.find('span', class_="sku").text
    i_kod = int(i[5:len(i)])
    
    return i_kod
```

```python
#Тип упаковки

def get_pack(link):
    ''' Переходим на страницу конкретного товара и ищем тип упаковки товара'''

    r = requests.get(link, HEADERS)
    soup = BeautifulSoup(r.text, 'lxml')
    pack = soup.find('span', class_="type-packaging-name").text
    
    return pack
```

**готовим ссылки с доп параметрами для переход на другую страницу списка**

```python
for i in range(1, 10):
    link = f'https://smallpacking.agrosem.ua/shop/page/{i}/'
    print(link)
```

**создаем базу данных**

```python
import sqlite3

connection = sqlite3.connect('product_list.db')
cursor = connection.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS product_list (Title TEXT, pack_type TEXT, KSU INT, Price INT)''')

connection.commit()
```

**собираем все воедино**

```python
for i in range(1, 9):
    SITE = f'https://smallpacking.agrosem.ua/shop/page/{i}/'
    r = requests.get(SITE, HEADERS)
    if r.status_code == 200:
        soup = BeautifulSoup(r.text, 'lxml')
        product = soup.find('div', class_="product-from-category-container")
        for t in product.parent.find_all('div', class_="product-from-category-container"):
            name = t.span.text
            link = t.a.get('href')
            i_kod = get_i_kod(link)
            price = get_price(link)
            pack = get_pack(link)
            cursor.execute("INSERT INTO product_list VALUES (?,?,?,?)", (name, pack, i_kod, price))
            print(name, pack, i_kod, price)
```

```python
for i in cursor.execute('SELECT * FROM product_list'):
    print(i)
```
